### Limits of Diversification
(including the anatomy of a cap-weighted portfolio, and a bunch of cool Pandas stuff along the way)

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import edhec_risk_kit as erk
import numpy as np
import pandas as pd

ind_return = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

```python
ind_return.shape
```

```python
ind_nfirms.shape
```

```python
ind_size.shape
```

```python
ind_nfirms.head()
```

```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis='columns')
```

```python
total_mktcap.plot(figsize=(12,6))
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis='rows')
```

```python
ind_capweight.head()
```

```python
# make sure the cap weights added up to 100%
ind_capweight['1926'].sum(axis='columns')
```

```python
ind_capweight[['Fin', 'Steel']].plot(figsize=(12,6))
```

```python
total_market_return = (ind_capweight * ind_return).sum(axis='columns')
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(12,6), title='Total Market CapWeighted Index 1926-2018')
```

```python
total_market_index['1980':].plot(figsize=(12,6))
# also, we plot the 36-month moving average of the index by calling the method "rolling"
total_market_index['1980':].rolling(window=36).mean().plot(figsize=(12,6))
```

```python
tmi_tr36mrets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
tmi_tr36mrets.plot(figsize=(12,6), label="Trailing 36-month Returns", legend=True)
total_market_return.plot(label="Returns", legend=True)
```

### Rolling Correlations -  along with MultiIndexes and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.index.names = ['date', 'industry']
ts_corr.tail()
```

```python
'''
The problem with the above matrix is that it returns the correlation by date,
but what we really want is the correlation of the entire time series. Let's fix it
'''
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36mrets.plot(figsize=(12,6), label='Tr36 Monthly Returns', legend=True)
ind_tr36corr.plot(label='Tr36 Monthly Correlations', legend=True, secondary_y=True)
```

```python
'''
According to the above chart, when the market return is tanking, the correlation goes up, meaning most industries
goes down alongside the market.
This infers that diversification fails when the entire market is falling apart
'''
# let's zoom in. Plot the market after 2007
tmi_tr36mrets['2007':].plot(figsize=(12,6), label='Tr36 Monthly Returns', legend=True)
ind_tr36corr['2007':].plot(label='Tr36 Monthly Correlations', legend=True, secondary_y=True)
```

```python
# calculate the correlation between trailing 36-month returns and trailing 36-month correlations
tmi_tr36mrets.corr(ind_tr36corr)
```
