```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import edhec_risk_kit as erk
import numpy as np
import pandas as pd

ind = erk.get_ind_returns()
er = erk.annualize_rets(ind['1996':'2000'], 12)
cov = ind['1996':'2000'].cov()
```

```python
l = ['Food', 'Steel']
erk.msr(0.1, er[l], cov.loc[l,l])
```

```python
# let's experiment: twist the individual expected return and test how this impact the porfolio weights
erk.msr(0.1, np.array([0.11, 0.12]), cov.loc[l,l])
```

```python
erk.msr(0.1, np.array([0.13, 0.1]), cov.loc[l,l])
```

```python
erk.msr(0.1, np.array([0.09, 0.14]), cov.loc[l,l])
```

```python
erk.plot_ef(20, er, cov, show_cml=True, riskfree_rate=0.1)
```

```python
# plot the Equally Weighted porfolios, aka the naive portfolio
erk.plot_ef(20, er, cov, show_cml=True, riskfree_rate=0.1, show_ew=True)
```

```python
# plot the Global Minimum Volatility porfolios
erk.plot_ef(20, er, cov, show_cml=True, riskfree_rate=0.1, show_ew=True, show_gmv=True)
```
